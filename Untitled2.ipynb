```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cdist
```

```python
train_data = pd.read_csv('fashion-mnist_train.csv')

print("Dataset Shape:", train_data.shape)
train_data.head()
```

```python
label_names = {
    0: 'T-shirt/top', 1: 'Trouser', 2: 'Pullover', 3: 'Dress',
    4: 'Coat', 5: 'Sandal', 6: 'Shirt', 7: 'Sneaker',
    8: 'Bag', 9: 'Ankle boot'
}

fig, axes = plt.subplots(2, 5, figsize=(12, 5))

for i, ax in enumerate(axes.flat):
    image = train_data.iloc[i, 1:].values.reshape(28, 28)
    label = train_data.iloc[i, 0]

    ax.imshow(image, cmap='gray')
    ax.set_title(label_names[label])
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
sample_size = 1000

X = train_data.iloc[:sample_size, 1:].values
y = train_data.iloc[:sample_size, 0].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=50)
embeddings = pca.fit_transform(X_scaled)

print("Embeddings shape:", embeddings.shape)
```

```python
from scipy.spatial.distance import cdist
import numpy as np

# Compute inverse covariance matrix
VI = np.linalg.inv(np.cov(embeddings.T))

test_indices = [50, 150, 250, 350, 450]

print("SIMILARITY ANALYSIS - MAHALANOBIS\n")
print("="*80)

for idx in test_indices:
    new_product_embedding = embeddings[idx].reshape(1, -1)
    new_product_label = y[idx]

    # Mahalanobis distance
    distances = cdist(new_product_embedding, embeddings, metric='mahalanobis', VI=VI)[0]

    # Convert to similarity
    similarities = 1 / (1 + distances)

    # Top 3 similar (excluding itself)
    similar_indices = similarities.argsort()[-4:-1][::-1]

    print(f"\nNEW PRODUCT #{idx}: {label_names[new_product_label]}")
    print("Most similar past products:")

    for i, sim_idx in enumerate(similar_indices, 1):
        similarity_score = similarities[sim_idx]
        similar_label = y[sim_idx]
        print(f"  {i}. Product #{sim_idx}: {label_names[similar_label]} (Similarity: {similarity_score:.3f})")

    print("-"*80)
```

```python
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(embeddings)

plt.figure(figsize=(8, 6))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=y, cmap='tab10')
plt.title("t-SNE Clustering of Fashion Products")
plt.show()
```

```python
def compute_all_similarities(new_vec, embeddings):

    cos = cosine_similarity(new_vec, embeddings)[0]

    euc = 1 / (1 + pairwise_distances(new_vec, embeddings, metric='euclidean')[0])

    man = 1 / (1 + pairwise_distances(new_vec, embeddings, metric='manhattan')[0])

    VI = np.linalg.inv(np.cov(embeddings.T))
    maha = 1 / (1 + cdist(new_vec, embeddings, metric='mahalanobis', VI=VI)[0])

    cheb = 1 / (1 + pairwise_distances(new_vec, embeddings, metric='chebyshev')[0])

    corr = 1 / (1 + pairwise_distances(new_vec, embeddings, metric='correlation')[0])

    canb = 1 / (1 + pairwise_distances(new_vec, embeddings, metric='canberra')[0])

    return {
        "Cosine": cos,
        "Euclidean": euc,
        "Manhattan": man,
        "Mahalanobis": maha,
        "Chebyshev": cheb,
        "Correlation": corr,
        "Canberra": canb
    }
```

```python
test_indices = [50, 150, 250, 350, 450]

results = []

for idx in test_indices:
    new_vec = embeddings[idx].reshape(1, -1)
    true_label = y[idx]

    sims = compute_all_similarities(new_vec, embeddings)

    for method, sim in sims.items():
        top_idx = sim.argsort()[-4:-1][::-1]
        predicted_labels = y[top_idx]

        accuracy = np.mean(predicted_labels == true_label)

        results.append([idx, method, accuracy])
```

```python
results_df = pd.DataFrame(results, columns=["Product", "Method", "Accuracy"])

results_df
```

```python
avg_results = results_df.groupby("Method")["Accuracy"].mean()

print(avg_results)
```

```python
plt.figure()

avg_results.plot(kind='bar')

plt.title("Comparison of Similarity Methods")
plt.xlabel("Method")
plt.ylabel("Accuracy")
plt.xticks(rotation=45)

plt.show()
```

# Final Demand Prediction using Best Similarity Method


```python
# -------------------------------
# Simulate sales data (since dataset doesn't have real sales)
# -------------------------------
np.random.seed(42)
sales_old = np.random.randint(100, 1000, size=sample_size)

# -------------------------------
# Pick a new product
# -------------------------------
test_idx = 100
new_vec = embeddings[test_idx].reshape(1, -1)

# -------------------------------
# Compute Mahalanobis similarity
# -------------------------------
VI = np.linalg.inv(np.cov(embeddings.T))
maha_sim = 1 / (1 + cdist(new_vec, embeddings, metric='mahalanobis', VI=VI)[0])

# -------------------------------
# Get top 5 similar products
# -------------------------------
top_idx = maha_sim.argsort()[-6:-1][::-1]

sim_k = maha_sim[top_idx]
sales_k = sales_old[top_idx]

# -------------------------------
# Demand Prediction Formula
# -------------------------------
predicted_demand = np.sum(sim_k * sales_k) / np.sum(sim_k)

print("Top similar product indices:", top_idx)
print("Their sales:", sales_k)
print("Predicted Demand:", int(predicted_demand))
```
