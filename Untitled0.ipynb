```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the training data
train_data = pd.read_csv('fashion-mnist_train.csv')

# Check what we have
print(f"Dataset shape: {train_data.shape}")
print(f"Columns: {train_data.columns.tolist()}")
print(train_data.head())
```

```python
# Create a label mapping
label_names = {
    0: 'T-shirt/top',
    1: 'Trouser',
    2: 'Pullover',
    3: 'Dress',
    4: 'Coat',
    5: 'Sandal',
    6: 'Shirt',
    7: 'Sneaker',
    8: 'Bag',
    9: 'Ankle boot'
}

# Display 10 sample images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    # Get image pixels (all columns except 'label')
    image = train_data.iloc[i, 1:].values.reshape(28, 28)
    label = train_data.iloc[i, 0]

    ax.imshow(image, cmap='gray')
    ax.set_title(f'{label_names[label]}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Take first 1000 images to work with (faster)
sample_size = 1000
X = train_data.iloc[:sample_size, 1:].values  # Get pixel values
y = train_data.iloc[:sample_size, 0].values   # Get labels

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce to 50 dimensions using PCA (this is creating embeddings)
pca = PCA(n_components=50)
embeddings = pca.fit_transform(X_scaled)

print(f"Created embeddings with shape: {embeddings.shape}")
print("Each image is now represented by 50 numbers instead of 784 pixels")
```

```python
from sklearn.manifold import TSNE

# Reduce 50 dimensions to 2 dimensions so we can plot it
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(embeddings)

# Plot the embeddings, colored by clothing type
plt.figure(figsize=(12, 10))
scatter = plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1],
                     c=y, cmap='tab10', alpha=0.6, s=50)

# Add legend
handles, labels = scatter.legend_elements()
legend_labels = [label_names[int(label.split('{')[1].split('}')[0])] for label in labels]
plt.legend(handles, legend_labels, title="Product Type", bbox_to_anchor=(1.15, 1))

plt.title('Fashion Product Embeddings - Similar Products Cluster Together', fontsize=14)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Pick 5 random "new" products to test
test_indices = [50, 150, 250, 350, 450]

print("SIMILARITY ANALYSIS - Comparing New Products to Past Products\n")
print("="*80)

for idx in test_indices:
    # Get the "new" product
    new_product_embedding = embeddings[idx].reshape(1, -1)
    new_product_label = y[idx]

    # Calculate similarity with all other products
    similarities = cosine_similarity(new_product_embedding, embeddings)[0]

    # Find top 3 most similar products (excluding itself)
    similar_indices = similarities.argsort()[-4:-1][::-1]

    print(f"\nNEW PRODUCT #{idx}: {label_names[new_product_label]}")
    print(f"Most similar past products:")
    for i, sim_idx in enumerate(similar_indices, 1):
        similarity_score = similarities[sim_idx]
        similar_label = y[sim_idx]
        print(f"  {i}. Product #{sim_idx}: {label_names[similar_label]} (Similarity: {similarity_score:.3f})")
    print("-"*80)
```
